## Feature Engineering & Feature Selection

```python
#extract Review Month from Review Effective Date
policy_review_df['EFFECTIVE_DATE'] = pd.DatetimeIndex(policy_review_df['EFFECTIVE_DATE'].values)
policy_review_df['REVIEW_MONTH'] = policy_review_df['EFFECTIVE_DATE'].dt.month
policy_review_df = policy_review_df.drop(columns=['EFFECTIVE_DATE'])
```

```python
policy_review_df.shape
```

```python
policy_review_df.info()
```

```python
policy_review_df['PRIMARY_AGENT_STATUS'].unique
```

```python
policy_review_df = policy_review_df.drop(policy_review_df.iloc[:,18:47],axis=1)
policy_review_df.info()
```

```python
policy_review_df['PREMIUM_INCREASE_5Y'] = policy_review_df['OPTIMUM_PREMIUM_5Y'] - policy_review_df['PREMIUM_BEFORE']
```

```python
policy_review_df = policy_review_df.dropna()
```

```python
policy_review_df.info()
```

```python
policy_review_desc = policy_review_df.describe()
policy_review_desc.to_csv("Policy Review Data - Statistical Analysis.csv", index=False)
```

```python
high_level_outcomes_dict = {'Option 1 - 5 year increase': 'REMAIN',
                   'Option 1a - Other Premium Increase': 'REMAIN',
                    'Option 2 - 10 year increase': 'REMAIN',
                    'Option 3 - Maintain current premium and reduce benefits': 'REMAIN',
                    'Option 3a - Reduce Premium and Benefits': 'REMAIN',
                    'Option 4 - Cover allowed to expire': 'LAPSE',
                     'Lapsed Immediately': 'LAPSE',
                    }
high_level_outcomes_dict
```

```python
policy_review_df['NEW_REVIEW_OUTCOME'] = policy_review_df['REVIEW_OUTCOME'].map(high_level_outcomes_dict)
policy_review_df = policy_review_df.drop(columns = ['REVIEW_OUTCOME'])
policy_review_df = policy_review_df.rename(columns = {"NEW_REVIEW_OUTCOME":'REVIEW_OUTCOME'})
policy_review_df.head()
```

### Encode Categorical Data

```python
from sklearn.preprocessing import OrdinalEncoder

ord_enc = OrdinalEncoder()
policy_review_df[["CRITICAL_ILLNESS_BEFORE","BRANCH","PRIMARY_AGENT_STATUS","ASSIGNED_FLAG","ASTERISKED_ORPHAN","REVIEW_OUTCOME","SINGLE","JOINT_LIFE_1ST", "DUAL_LIFE"]] = ord_enc.fit_transform(policy_review_df[["CRITICAL_ILLNESS_BEFORE","BRANCH","PRIMARY_AGENT_STATUS","ASSIGNED_FLAG","ASTERISKED_ORPHAN","REVIEW_OUTCOME","SINGLE","JOINT_LIFE_1ST", "DUAL_LIFE"]])
policy_review_df.head()
```

```python
policy_review_df['REVIEW_OUTCOME'] = policy_review_df['REVIEW_OUTCOME'].replace(1,2)
policy_review_df['REVIEW_OUTCOME'] = policy_review_df['REVIEW_OUTCOME'].replace(0,1)
policy_review_df['REVIEW_OUTCOME'] = policy_review_df['REVIEW_OUTCOME'].replace(2,0)
policy_review_df.head()
```

```python
# Visualizing a Pandas Correlation Matrix Using Seaborn
from matplotlib.pyplot import figure
figure(figsize=(12, 10), dpi=80)
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

matrix = policy_review_df.corr().round(2)
sns.heatmap(matrix, annot=True)
plt.xticks(rotation = 75)
plt.show()
```
